```python
# Fleur-de-lis is a lipstick company, a reputable name in the beauty industry. Our products let your inner charm shine and give you a royal look 
# so you can go out and conquer the world. With 5 different aesthetics, our lipsticks are fit for any occasion.
# Company owned by Lelia Erscoi

class Lipstick():

  def __init__(self):

    # The intrinsic characteristics of the product
    self.name                    = None              # The name of the product
    self.color                   = None              # What color the lipstick is
    self.ingredient_list         = None              # The list of ingredients contained
    self.finish                  = None              # How the lipstick will settle
    self.effect                  = None              # The desired look after use
    self.price                   = None              # The cost of the lipstick
    self.inventory               = None              # How many units of one lipstick are in stock

    self.active_ingredient1     = None
    self.active_ingredient2     = None
    self.active_ingredient3     = None
    
  def data_extracting(self, product_information):
    self.name            = product_information.split(' - ')[0]
    self.color           = product_information.split(' - ')[1]
    self.ingredient_list = product_information.split(' - ')[2]
    self.finish          = product_information.split(' - ')[3]
    self.effect          = product_information.split(' - ')[4]
    self.price           = product_information.split(' - ')[5]
    self.inventory       = product_information.split(' - ')[-1]

  def data_featuring(self):
    self.active_ingredient1 = self.ingredient_list.split(', ')[0]
    self.active_ingredient2 = self.ingredient_list.split(', ')[1]
    self.active_ingredient3 = self.ingredient_list.split(', ')[-1]


  def return_data_cleaned(self):
    return({
        "Lipstick name"        : self.name,
        "Color"                : self.color,
        "Ingredients"          : self.ingredient_list,
        "Finish"               : self.finish,
        "Effect"               : self.effect,
        "Price (euros)"        : self.price,
        "Inventory"            : self.inventory
    })

  # Changes the price of the lipstick
  def update_price(self, new_price):
    self.price += new_price

  # Checks the ingredient list for animal based ingredients 
  def is_vegan(self):
    animal_ingredients = ["carmine", "beeswax", "lanolin_oil"]
    check_vegan = ([ingredient for ingredient in animal_ingredients if(ingredient in self.ingredient_list)])
    if len(check_vegan) == 0:
      print("No worries, the {} product is vegan".format(self.name))
    else:
      print("The {} lipstick is not vegan because it contains {}".format(self.name, check_vegan))

  # Checks if lipstick is still available to be sold
  def in_stock(self):
    return "Yes, we still have this in stock!" if int(self.inventory) > 0 else "Sorry, product is out of stock."
 
  # Changes the price based on currency
  def currency_exchange(self, currency):
    exchange = CurrencyRates()
    rate = exchange.get_rate('EUR', currency) 
    return int(self.price) * rate
```

```python
class Client():

  def __init__(self):
    self.name                     = None
    self.age                      = None
    self.gender                   = None
    self.address                  = None
    self.order_time               = None
    self.delivery_time            = None
    self.add_preference           = None
    self.order_history            = None

  def data_extracting(self, client_info):
    client_information = str(client_info)
    self.name = client_information.split('~')[0]
    self.age = client_information.split('~')[1]
    self.address = client_information.split('~')[2]
    self.order_time = client_information.split('~')[3]
    self.add_preference = client_information.split('~')[-1]

  def data_featuring(self):
    self.gender   = 'M' if self.name.split(" ")[0] == 'Mr.' else 'F' if self.name.split(" ")[0] == 'Ms.' else "Other"
    self.order_history = []

  def return_data_cleaned(self):
    return({
        "Client Name"               : self.name,
        "Client Age"                : self.age,
        "Client Gender"             : self.gender,
        "Delivery Address"          : self.address,
        "Order Date"                : self.order_time,
        "Add Preferance"            : self.add_preference,
        "Order History"             : self.order_history
    })

  def estimated_delivery_time(self):
    dummy = datetime.datetime.strptime(self.order_time,' %Y-%m-%d %H:%M:%S ').date()
    self.delivery_time = dummy + timedelta(days=3)
    print("{}'s order ordered at {} will be delived on {}".format(self.name, self.order_time, self.delivery_time))

  def set_add_prefference(self, new_preference):
    self.add_preference = new_preference

  def new_order(self, lipstick):
    for item in lipstick:
      self.order_history.append(item.name)
    return self.order_history

  def orders_so_far(self):
    print("Client {} has ordered {} items so far.".format(self.name,len(self.order_history)))
```
