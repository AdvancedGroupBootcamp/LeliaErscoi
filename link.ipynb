### Cleaning data


```python
# renaming our columns
d = d.rename({0: 'Client 1', 1: 'Client 2'}, axis='columns')
d.head(1)
```

```python
# Since we only have 2 clients, rows with NaN values are useless, so we will get rid of them
d.dropna(subset=['Client 1', 'Client 2'], inplace=True)
d
```

```python
# Let's reset the index to access our data better
d = d.reset_index()
d.columns = ['Information', 'Client 1', 'Client 2']
```

```python
# We can already see that the snapshot rows can be formatted better
snapshot_rows = d[d['Information'].str.contains('snapshot')]['Information']
snapshot_rows_clean = snapshot_rows.apply(lambda x: '_'.join(x.split("snapshot.")[1].split("."))).tolist()
col_rename_dict = {i:j for i,j in zip(snapshot_rows, snapshot_rows_clean)}

list_ind = 0
for i in d.index: 
  if d.at[i, 'Information'] in snapshot_rows.tolist():
    d.at[i, 'Information'] = snapshot_rows_clean[list_ind]
    list_ind += 1  
```

```python
d
```

```python
# Does the same happen for cart?
cart_rows = d[d['Information'].str.contains('cart')]['Information']
cart_rows_clean = cart_rows.apply(lambda x: '_'.join(x.split("cart.")[1].split("."))).tolist()
col_rename_dict = {i:j for i,j in zip(cart_rows, cart_rows_clean)}

list_ind = 0
for i in d.index: 
  if d.at[i, 'Information'] in cart_rows.tolist():
    d.at[i, 'Information'] = cart_rows_clean[list_ind]
    list_ind += 1  
```

```python
# For easier analysis we will convert all bools to either 0 or 1
d = d *1
d
```

```python
# Get rid of id rows as we don't have other df to reference to
ids = [d['Information'].tolist().index(c) for c in d['Information'].tolist() if "id" in c]  + [d['Information'].tolist().index(c) for c in d['Information'].tolist() if "Id" in c] 
ids 
```

```python
d = d.drop(labels= ids, axis=0)
d.reset_index(drop=True) # some values will be skipped in the index to we adjust that
d
```

```python
# Some sketchy things happening with the phone numbers...
display(d.loc[d['Information'] == 'patientPhoneNumber'])

phones_raw1 = d.loc[d['Information'] == 'patientPhoneNumber']['Client 1'].to_dict()[26].replace('\'',"")
phones_raw2 = d.loc[d['Information'] == 'patientPhoneNumber']['Client 2'].to_dict()[26].replace('\'',"")

phone_number1 = ''.join(re.findall(r'\d+', phones_raw1))
phone_number2 = ''.join(re.findall(r'\d+', phones_raw2))

d.loc[d.Information == 'patientPhoneNumber', ['Client 1','Client 2']] = phone_number1 , phone_number2
d
```

```python
from nltk import tokenize
tokenizer = tokenize.SpaceTokenizer()
```

```python
# Making the details column easier to work with
client1_simptoms = d.loc[d['Information'] == 'details']['Client 1'].apply(lambda s: tokenizer.tokenize(s))
client2_simptoms = d.loc[d['Information'] == 'details']['Client 2'].apply(lambda s: tokenizer.tokenize(s))

client1_simptoms_clean = [re.sub(r'[^A-Za-z0-9]+', ',', str(x)) for x in client1_simptoms]
client2_simptoms_clean = [re.sub(r'[^A-Za-z0-9]+', ',', str(x)) for x in client2_simptoms]

d.loc[d.Information == 'details', ['Client 1','Client 2']] = client1_simptoms_clean[0], client2_simptoms_clean[0]
d
```

```python
# Administrative area of client2 has a typo!
area = [re.sub(r'[^0-9]+', '', str(x)) for x in d.loc[d['Information'] == 'patientAdministrativeArea']['Client 2']]
d.loc[d.Information == 'patientAdministrativeArea', 'Client 2'] = area
d
```

```python
# Extractint date and time from updated at
date1, time1 = d.loc[d['Information'] == 'updatedAt']['Client 1'].to_string().split(" ")[-1].split("T")
date2, time2 = d.loc[d['Information'] == 'updatedAt']['Client 2'].to_string().split(" ")[-1].split("T")
updatedAtDate = {'Information':'updatedAtDate', 'Client 1': date1, 'Client 2': date2}
updatedAtTime = {'Information':'updatedAtTime', 'Client 1': time1, 'Client 2': time2}
d = d.append([updatedAtDate, updatedAtTime], ignore_index=False)
d
```

```python
# One last very well hidden dict to extract!
dict_to_extract = d.loc[d['Information'] == 'paymentIntents']['Client 2'].to_dict()[55][0]
dict_to_extract
```

```python
# Let's make another dataframe this time, it will be easy to connect them!
payments = pd.DataFrame(dict_to_extract.items(),columns = ['Key','Value']) 
filter = payments["Value"] != ""
paymentsNew = payments[filter]
paymentsNew
```
