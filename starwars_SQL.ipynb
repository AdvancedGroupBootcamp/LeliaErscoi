# SQL Analysis

####  Q1: How many women my age are there in starwars?

```python
pd.read_sql(
    '''
        SELECT 
            name
        FROM 
            People
        WHERE
            gender = 'female'
        AND
            age BETWEEN 20 AND 25
    ''', 
    con
)
```

#### Q2: How many planets of temperate climate are there?

```python
pd.read_sql(
    '''
        SELECT 
            climate,
            COUNT (*) as amount
        FROM 
            Planets
        WHERE
            climate LIKE '%temperate%'
    ''', 
    con
)
```

### Inner
#### I want a friend that never dies, who should I choose?

```python
pd.read_sql(
    '''
        SELECT 
            p.name AS name,
            s.classification AS classification,
            s.designation AS designation,
            s.average_lifespan AS lifespan
        FROM 
            People p 
        INNER JOIN 
            Species s ON  p.species == s.name 
        WHERE
          s.classification = 'artificial'
        OR
          s.average_lifespan > 100000
   
    ''', 
    con
)
```

### Left
#### Who are the characters who come from the most populated planets?

```python
pd.read_sql(
    '''
        SELECT 
            p.name AS name,
            plan.name AS planet,
            plan.population AS population
        FROM 
            People p
        LEFT JOIN 
            Planets plan ON  p.homeworld  = plan.name 
        WHERE 
            planet != "None"
        ORDER BY
            plan.population DESC
        LIMIT 5
    ''', 
    con
)
```

### Group By
#### What are the smallest episodic characters?

```python
pd.read_sql(
    '''
        SELECT 
            f.title,
            p.name,
            p.height as height
        FROM 
            Films f
        INNER JOIN
            People p
        ON
          f.title = p.films
        GROUP BY
          p.height
        LIMIT 7
    ''', 
    con
)
```

### Having
#### What are the fastest vehicles we can get for the smallest amount of money?

```python
pd.read_sql(
    '''
        SELECT 
            *
        FROM 
            Vehicles
        WHERE 
            cost_in_credits != 0
        GROUP BY
            max_atmosphering_speed 
        HAVING
            cost_in_credits < 15000
    ''', 
    con
)
```

### CTE
#### Which characters are pushing their luck? Looking for older than average characters that drive super fast ships

```python
pd.read_sql(
    '''
        WITH
          average_age AS(
            SELECT
              p.name AS character_name,
              s.name AS species 
            FROM
              People p
              INNER JOIN Species s ON p.species = s.name
              WHERE p.age > s.average_lifespan
          ),

          fast AS(
            SELECT 
              p.name AS character_name,
              s.name AS ship 
            FROM
              People p
              INNER JOIN Starships s ON p.starships = s.name
              WHERE s.max_atmosphering_speed > 787.5
          )

          SELECT 
            *
          FROM 
            average_age

          UNION

          SELECT
            *
          FROM 
            fast
    ''', 
    con
)
```
